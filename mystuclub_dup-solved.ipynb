```python
import pandas as pd
```

```python
df = pd.read_json('mystudentdup.json')
```

```python
df.head(10)
```

```python
df.info()
df.describe(include='all')
```

```python
total_rows = len(df)
unique_ids = df['id'].nunique()

print("Total rows:", total_rows)
print("Unique job IDs:", unique_ids)
print("Duplicate count:", total_rows - unique_ids)
```

```python
# Find all duplicated IDs
duplicate_ids = df[df['id'].duplicated(keep=False)].sort_values('id')

duplicate_ids
```

```python
df['id'].value_counts()[df['id'].value_counts() > 1]
```

```python
import matplotlib.pyplot as plt

df['id'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Frequency of Job IDs (Duplicates highlighted)")
plt.xlabel("Job ID")
plt.ylabel("Count")
plt.show()
```

```python
df_clean = df.drop_duplicates(subset='id', keep='last')
```

```python
print("Before:", df.shape)
print("After:", df_clean.shape)
```

```python
df_clean.to_json("cleaned_jobs.json", orient="records", indent=4)
```
